```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as pld
from sklearn.linear_model import LinearRegression
```

```python
data = pandas.read_csv('cost-revenue-clean.csv')
```

```python
data.describe()
```

```python
3.29e7
```

```python
X = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
pld.figure(figsize= (10,6))
pld.scatter(X,y, alpha=0.3)
pld.title('Film cost vs Global revenue')
pld.xlabel('Production Budget $')
pld.ylabel('Worldwide Gross $')
pld.ylim(0,3000000000)
pld.xlim(0,450000000)
pld.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

SlopeCoefficient


```python
regression.coef_
```

```python
regression.intercept_
```

```python
pld.figure(figsize= (10,6))
pld.scatter(X,y, alpha=0.3)

#Adding regression line here
pld.plot(X, regression.predict(X), color= 'Red', linewidth=3)
pld.title('Film cost vs Global revenue')
pld.xlabel('Production Budget $')
pld.ylabel('Worldwide Gross $')
pld.ylim(0,3000000000)
pld.xlim(0,450000000)
pld.show()
```

```python
#getting R square from regression
regression.score(X, y)
```

R-square is 0.5496
